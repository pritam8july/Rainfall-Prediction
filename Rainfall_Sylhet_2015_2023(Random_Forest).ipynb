#Importing Dataset

```python
filepath = "sylhet_data_v_4.xlsx"
dataset = pd.read_excel(filepath)
x = dataset.iloc[ : , 1:21].values
y = dataset.iloc[:,-1].values
dataset.head(3)
```

```python
print(x)
```

```python
print(y)
```

```python
y = y.reshape(-1,1) # convert 1D arry to 2D array
```

```python
print(y)
```

#Taking Care of Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
x = imputer.fit_transform(x)
y = imputer.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

#Encoding the Dataset

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
x[:,10] = le1.fit_transform(x[:,10])
le2 = LabelEncoder()
x[:,-1] = le2.fit_transform(x[:,-1])
le3 = LabelEncoder()
y = le3.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
print(x)
```

#Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

#Training Model

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=150, random_state=0)
classifier.fit(x_train, y_train)
```

```python
classifier.score(x_train,y_train)
```

```python
y_pred = le3.inverse_transform(np.array(classifier.predict(x_test),dtype=int))
y_test = le3.inverse_transform(np.array(y_test,dtype=int))
```

```python
# print(y_pred)
```

```python
# print(y_test)
```

```python
y_pred = y_pred.reshape(-1,1)
y_test = y_test.reshape(-1,1)
# print(y_pred)
```

```python
# print(x_test)
```

```python
df = np.concatenate((y_test,y_pred),axis=1)
dataframe = pd.DataFrame(df,columns=['Rain on Tommorrow','Predition of Rain'])
print(dataframe)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

#Visualizing the Results

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```
